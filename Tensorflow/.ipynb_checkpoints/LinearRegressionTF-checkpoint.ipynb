```python
# make a simple linear regression predictor that tries to minimize the error
# in the actual and predicted values by using a mean square loss function
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# initialize our data with 50 random points
X = np.random.rand(50)
Y = X * 2 + 5
plt.plot(X, Y)
plt.show()
```

```python
# now we can initialize the tensorflow variables to use for the regression
# they act as our initial prediction for our line and we need to improve this
weight = tf.Variable(tf.random_uniform([1], -5.0, 5.0))
bias = tf.Variable(tf.zeros(1))
predicted_values = weight * X + bias
loss_function = tf.reduce_mean(tf.square(Y - predicted_values))
optimizer = tf.train.GradientDescentOptimizer(0.2)  # our learning rate
train_func = optimizer.minimize(loss_function)

with tf.Session() as session:
    init = tf.global_variables_initializer()
    session.run(init)
    for iteration in range(100):
        session.run(train_func)
        if iteration % 10 == 0:
            print('Currently done with iteration', iteration)
            print('Currently predicted value of weight is', session.run(weight),
                   'and bias', session.run(bias))
    fig1, axes_array = plt.subplots(1, 2)
    axes_array[0].plot(X, Y)
    axes_array[0].set(title='Original Function')
    axes_array[1].plot(X, X * session.run(weight)[0] + session.run(bias)[0])
    axes_array[1].set(title='Predicted Function')
    plt.show()
```
